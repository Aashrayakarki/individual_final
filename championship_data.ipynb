```python
import pandas as pd
from unidecode import unidecode
import requests
from io import StringIO
```

```python
url = 'https://fbref.com/en/comps/10/2023-2024/defense/2023-2024-Championship-Stats'
html_content = requests.get(url).text.replace('<!--', '').replace('-->', '')
df = pd.read_html(StringIO(html_content))
```

```python
df[2].columns=df[2].columns.droplevel(0)
```

```python
dfdata = df[2]
```

```python
dfdata.head()
```

```python
len(dfdata[dfdata.Age=='Age'])
```

```python
dfcreation=dfdata.drop(dfdata[dfdata.Age=='Age'].index)
```

```python
dfcreation
```

```python
dfcreation.columns
```

```python
df_after_column_removal=dfcreation.drop(columns=['Sh', 'Pass', 'Err', 'Matches'])
```

```python
df_after_column_removal['90s'] = df_after_column_removal['90s'].astype(float)
df_after_column_removal['Tkl'] = df_after_column_removal['Tkl'].astype(float)
df_after_column_removal['TklW'] = df_after_column_removal['TklW'].astype(float)
df_after_column_removal['Def 3rd'] = df_after_column_removal['Def 3rd'].astype(float)
df_after_column_removal['Mid 3rd'] = df_after_column_removal['Mid 3rd'].astype(float)
df_after_column_removal['Att 3rd'] = df_after_column_removal['Att 3rd'].astype(float)
df_after_column_removal['Blocks'] = df_after_column_removal['Blocks'].astype(float)
df_after_column_removal['Int'] = df_after_column_removal['Int'].astype(float)
df_after_column_removal['Tkl+Int'] = df_after_column_removal['Tkl+Int'].astype(float)
df_after_column_removal['Clr'] = df_after_column_removal['Clr'].astype(float)
```

```python
df_after_column_removal.head()
```

```python
finaldf = df_after_column_removal
```

```python
finaldf['Tkl']=finaldf['Tkl']/finaldf['90s']
finaldf['TklW']=finaldf['TklW']/finaldf['90s']
finaldf['Def 3rd']=finaldf['Def 3rd']/finaldf['90s']
finaldf['Mid 3rd']=finaldf['Mid 3rd']/finaldf['90s']
finaldf['Att 3rd']=finaldf['Att 3rd']/finaldf['90s']
finaldf['Blocks']=finaldf['Blocks']/finaldf['90s']
finaldf['Int']=finaldf['Int']/finaldf['90s']
finaldf['Tkl+Int']=finaldf['Tkl+Int']/finaldf['90s']
finaldf['Clr']=finaldf['Clr']/finaldf['90s']
```

```python
finaldf.head()
```

```python
finaldf['Player']=finaldf['Player'].apply(unidecode)
```

```python
finaldf['Squad']=finaldf['Squad'].apply(unidecode)
```

```python
finaldf.head()
```

```python
finaldf = finaldf[finaldf['Pos'] == 'DF']
```

```python
finaldf.loc[:, 'Age'] = pd.to_numeric(finaldf['Age'])
```

```python
finaldf1 = finaldf[finaldf['Age'] <= 23]
```
